## 📝 Redacción del Problema
El sistema debe presentar una interfaz de usuario mediante un menú interactivo que permita realizar cálculos de manera repetitiva sin cerrar el programa. El software debe gestionar la entrada de datos numéricos para la base y la altura, procesar estos valores a través de funciones especializadas para el área y el perímetro, y finalmente mostrar los resultados de forma clara. El programa debe incluir una opción de salida para finalizar la ejecución de forma controlada.

Para realizar los cálculos, el programa implementa las siguientes definiciones de la geometría euclidiana:Área ($A$): Es la medida de la superficie encerrada por los límites del rectángulo. Se obtiene multiplicando la base por la altura.$$A = b \cdot h$$Perímetro ($P$): Es la suma de las longitudes de todos los lados del rectángulo. Dado que los lados opuestos son iguales, la fórmula se simplifica al doble de la suma de la base y la altura.$$P = 2(b + h)$$

```python
# ============================================================================
# 🎓 PROGRAMA DE CÁLCULOS GEOMÉTRICOS - RECTÁNGULOS
# ============================================================================
# Descripción: Calcula área y perímetro de un rectángulo usando funciones
# Nivel: Preparatoria
# ============================================================================

# ============================================================================
# FUNCIÓN: arearectangulo()
# Fórmula: A = base * altura
# ============================================================================
def arearectangulo():
    print("\n" + "="*60)
    print("🟦 CÁLCULO DEL ÁREA DEL RECTÁNGULO")
    print("="*60)
    print("Fórmula: A = base × altura")
    print("-"*60)

    base = float(input("✏️  Ingresa el valor de la base (b): "))
    altura = float(input("✏️  Ingresa el valor de la altura (h): "))
    area = base * altura

    print("\n📊 RESULTADOS:")
    print("-"*60)
    print(f"   Base ingresada:    {base}")
    print(f"   Altura ingresada:  {altura}")
    print(f"   🎯 Área resultante: {area:.2f} unidades²")
    print("="*60)
    input("\n💡 Presiona ENTER para continuar...")

# ============================================================================
# FUNCIÓN: perimetrorectangulo()
# Fórmula: P = 2 * (base + altura)
# ============================================================================
def perimetrorectangulo():
    print("\n" + "="*60)
    print("🟧 CÁLCULO DEL PERÍMETRO DEL RECTÁNGULO")
    print("="*60)
    print("Fórmula: P = 2 × (base + altura)")
    print("-"*60)

    base = float(input("✏️  Ingresa el valor de la base (b): "))
    altura = float(input("✏️  Ingresa el valor de la altura (h): "))
    perimetro = 2 * (base + altura)

    print("\n📊 RESULTADOS:")
    print("-"*60)
    print(f"   Base ingresada:         {base}")
    print(f"   Altura ingresada:       {altura}")
    print(f"   🎯 Perímetro resultante: {perimetro:.2f} unidades")
    print("="*60)
    input("\n💡 Presiona ENTER para continuar...")

# ============================================================================
# FUNCIÓN: menu()
# Descripción: Controla el flujo con if, elif, else y while
# ============================================================================
def menu():
    ejecutando = True

    while ejecutando:
        print("\n" * 2)
        print("╔" + "═"*58 + "╗")
        print("║" + " "*15 + "🎓 GEOMETRÍA DEL RECTÁNGULO 🎓" + " "*13 + "║")
        print("╚" + "═"*58 + "╝")
        print()
        print("┌" + "─"*58 + "┐")
        print("│" + " "*22 + "MENÚ PRINCIPAL" + " "*22 + "│")
        print("├" + "─"*58 + "┤")
        print("│" + " "*58 + "│")
        print("│   🟦 1. Calcular ÁREA del rectángulo               │")
        print("│   🟧 2. Calcular PERÍMETRO del rectángulo          │")
        print("│   🚪 3. SALIR del programa                         │")
        print("│" + " "*58 + "│")
        print("└" + "─"*58 + "┘")
        print()

        opcion = input("👉 Selecciona una opción (1-3): ")

        if opcion == "1":
            arearectangulo()  # Llamada a la función de área
        elif opcion == "2":
            perimetrorectangulo()  # Llamada a la función de perímetrO
        elif opcion == "3":
            # Bloque de despedida elegante
            print("\n" + "✨"*30)
            print("👋 ¡Gracias por usar el programa!")
            print("🎓 ¡Hasta pronto, estudiante!")
            print("✨"*30 + "\n")
            ejecutando = False  # Rompe el ciclo while

        else:
            # Manejo de error para opciones no válidas
            print("\n⚠️  OPCIÓN NO VÁLIDA")
            print("💡 Por favor, selecciona 1, 2 o 3")
            print("-"*60)
            input("💡 Presiona ENTER para intentar de nuevo...")

# ============================================================================
# PROGRAMA PRINCIPAL (Llamada inicial)
# ============================================================================
if __name__ == "__main__":
    menu()
```
